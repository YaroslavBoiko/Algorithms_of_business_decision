## Testing my DBSCAN

# Description of the algorithm
 1. Construction begins with an arbitrary point that has not been used before.
 2. A request is made for points from the Îµ-neighborhood, and if there are enough points, a cluster is formed. 
 3. Otherwise, the point is marked as noise. (Note that this point can be found later in a sufficiently large Îµ-neighborhood of another point and belong to the cluster, rather than being noise.)
 4. If a point is in a dense part of a group, then its Îµ-environment will also be part of this cluster.
 5. Therefore, all points of the Îµ-neighborhood will also be added to the cluster, as well as the Îµ-neighborhood of these points, if they are too dense.
 
 This process continues until the tightly coupled cluster is completely found. After that, a request is made for a    new raw point, which is processed and, or there is a new cluster, or it is marked as noise.

```python
import numpy
 
def MyDBSCAN(D, eps, min_PTS):
    
    def regionQuery(D, P, eps):
        '''
        The function that helps to find the nearest points (in D) around point P at a distance eps (epsilon).
        '''
        neighbors = []
        
        for Pn in range(0, len(D)):
        
            if numpy.linalg.norm(D[P] - D[Pn]) < eps:
                neighbors.append(Pn)
            
        return neighbors
            
    def growCluster(D, labels, P, NeighborPts, C, eps, MinPts):
        '''
        The function for building a cluster for the center.
        '''
        labels[P] = C
        i = 0
        while i < len(NeighborPts):
            Pn = NeighborPts[i]
            if labels[Pn] == -1:
                labels[Pn] = C
            
            elif labels[Pn] == 0:
                labels[Pn] = C
                
                PnNeighborPts = regionQuery(D, Pn, eps)
                
                if len(PnNeighborPts) >= MinPts:
                    NeighborPts = NeighborPts + PnNeighborPts
                    
            i += 1 
                
            
    # In this tutorial, we need to initialize the list of labels, which we will return as a result of the algorithm
    # First of all, we have not checked any point, so all of them now do not belong to any cluster
    labels_list = [0]*len(D) 
        
        
    Ñurr_claster = 0
    
    # The algorithm works in such a way that we need to check each point that does not belong to any cluster
    # Of course it is possible that a neighbor will have three neighbors that do not belong to the priority point, but this condition is checked by the function growCluster 
    for P in range(0, len(D)):
        
        
        # We need to check if the element belongs to a specific cluster. 
        # If it already applies, then we move on to the next point.
        # And if not then we continue to work with this to find a new cluster.
        if not (labels_list[P] == 0):
            continue
    
        # With the help of the function regionQuery we will find all possible neighbors that will form our cluster
        possible_neighbors = regionQuery(D, P, eps) 
        
        # Checking the number of neighbors to form a cluster

        if len(possible_neighbors) < min_PTS:
            labels_list[P] = -1
            
        else: 
            Ñurr_claster += 1
            #The process of creating a cluster Based on previous analysis and function growCluster
            growCluster(D, labels_list, P, possible_neighbors, Ñurr_claster, eps, min_PTS)
            
    return labels_list          
```
